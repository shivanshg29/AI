```python
from copy import deepcopy
def heuristic(s,g):
    d=0
    for i in range(0,len(s)):
        for j in range(0,len(s[0])):
            if(s[i][j]!=0):
                if s[i][j]!=g[i][j]:
                    d=d+1
    return d
```

```python
def findSpace(s):
    for i in range(0,len(s)):
        for j in range(0,len(s[0])):
            if s[i][j]==0:
                return (i,j)
```

```python
def up(s):
    i,j=findSpace(s)
    mat=deepcopy(s)
    mat[i][j],mat[i-1][j]=mat[i-1][j],mat[i][j]
    return mat

def down(s):
    i,j=findSpace(s)
    mat=deepcopy(s)
    mat[i][j],mat[i+1][j]=mat[i+1][j],mat[i][j]
    return mat
    
def left(s):
    i,j=findSpace(s)
    mat=deepcopy(s)
    mat[i][j],mat[i][j-1]=mat[i][j-1],mat[i][j]
    return mat

def right(s):
    i,j=findSpace(s)
    mat=deepcopy(s)
    mat[i][j],mat[i][j+1]=mat[i][j+1],mat[i][j]
    return mat
```

```python
def best_find(op):
    mn=99999
    ind=-1
    for i in range(0,len(op)):
        if mn>op[i][1]:
            mn=op[i][1]
            ind=i
    return ind
```

```python
def best_fs(s,g):
    op=[]
    closed=[]
    op.append([s,heuristic(s,g)])
    while len(op)!=0:
        ind=best_find(op)
        curr=op[ind][0]
        op.pop(ind)
        closed.append(curr)
        
        if curr==g:
            print("Goal Found")
            print(curr)
            break
        
        (row,col)=findSpace(curr)
        if row!=0:
            new_node=up(curr)
            print(new_node)
            if new_node not in op and new_node not in closed:
                op.append([new_node,heuristic(new_node,g)])
        
        if row!=2:
            new_node=down(curr)
            print(new_node)
            if new_node not in op and new_node not in closed:
                op.append([new_node,heuristic(new_node,g)])
        
        if col!=0:
            new_node=left(curr)
            print(new_node)
            if new_node not in op and new_node not in closed:
                op.append([new_node,heuristic(new_node,g)])
        
        if col!=2:
            new_node=right(curr)
            print(new_node)
            if new_node not in op and new_node not in closed:
                op.append([new_node,heuristic(new_node,g)])

        print("\n")
```

```python
def simpleHillClimb(s,g):
    curr=s
    while(True):
        if(curr==g):
            print("Goal Found")
            print(curr)
            break
        else:
            (row,col)=findSpace(curr)
            succ=[]
            if(row!=0):
                new_node=up(curr)
                h1=heuristic(new_node,g)
                succ.append([new_node,h1])
            if row!=2:
                new_node=down(curr)
                h2=heuristic(new_node,g)
                succ.append([new_node,h2])
            if col!=0:
                new_node=left(curr)
                h3=heuristic(new_node,g)
                succ.append([new_node,h3])
            if col!=2:
                new_node=right(curr)
                h4=heuristic(new_node,g)
                succ.append([new_node,h4])

            ind=best_find(succ)
            print(succ[ind][0],"\n")

            if (heuristic(curr,g)==succ[ind][1]):
                print("Local Maxima Reached")
                break

            curr=succ[ind][0]
```
